### Relevant Undergraduate Courses

Many graduate programs and data science jobs seek candidates who can demonstrate sufficient background in Computer Science, Math, and Statistics, as these skills will are important to suceeding in advanced computational activities. One way to demonstrate this is by taking college-level courses (in person or online) in the following subject areas:
* Computer Science
* Linear Algebra
* Calculus
* Statistics / Econometrics


### Free Online Material

#### Harvard University's CS 109 Data Science Course

All of the material, including lectures and assignments, are available for free online for Harvard University's two-part, year long Data Science course.

[**CS 109a: Introduction to Data Science**](https://harvard-iacs.github.io/2020-CS109A/): Introduces the analysis of data to perform predictions using statistical and machine learning methods. Topics include data scraping, data management, data visualization, regression and classification methods, and deep neural networks.

[**CS 109b: Advanced Topics in Data Science**](https://harvard-iacs.github.io/2020-CS109B/): Focuses on advanced methods for data wrangling, data visualization, statistical modeling, and prediction. Topics include big data, multiple deep learning architectures such as CNNs, RNNs, autoencoders, and generative models as well as basic Bayesian methods, nonlinear statistical models, and unsupervised learning. 

**Languages and Tools:** Python, Jupyter Notebooks/Google Colab

#### edX / Coursera

[edX](https://www.edx.org/course/subject/data-science) is a massive open online course (MOOC) provider founded by Harvard and MIT. Courses are taught by professors from these two universities and other institutions. They can be audited for free while MicroMasters and Professional Certificate Programs are also offered for a fee. 

[Coursera](https://www.coursera.org/browse/data-science) is another MOOC provider founded by Stanford University. Like edX, its course are typically taught by university faculty, and it offers free courses along with paid certifcate and specialization programs. 

Both platofrms have a wide range of class offerings at all levels: introductory statistics, introductory programming in various languages including Python, machine learning, etc.

#### Udemy

Udemy is another online course platform offering both free and paid courses, but it is not affiliated with any particular academic institution. Courses are not as academic in nature and are instead more technical. 

#### Towards Data Science

[**Towards Data Science**](https://towardsdatascience.com/) is a online publication on the Medium Ecosystem that publishes curated articles on data science. It's a great place to find  specific tutorials or articles discussing particular data science concepts. An account is required to access a limited number of free articles per month. Unlimited access is \$ 5/month or \$50/year. 

#### Stack Overflow

[Stack Overflow](https://stackoverflow.com/) is an online platform for programming questions and answers. It is one of the main resources for any programmer, including data scientists. Often, the top results of a Google search for a programming question will be a Stack Overflow post. 

#### Other Programming Languages
Although this workshop was taught in Python, there are many other programming languages used by data professionals. Generally, you only need one statistical programming language, such as Python or R, and basic SQL literacy.

**R**: R is another common free, open-source language used by data scientists. It was built by statisticians so its focus is statistics while Python is a more general programming language. It is commonly used along with a free graphical integrated development environment [RStudio](https://rstudio.com), available for Windows, macOS, and Linux. 

**SQL**: SQL (Structured Query Language) is a language used for managing data in relational databases. It can actually be used within Python and other programming languages. Because companies often hold data in such databases, it is almost always a required skill for a data scientist. 

**Julia**: An free, open-source alternative to Python, Julia was specifically designed for analysis and computational science. Thus, it can perform faster than Python.

**Master's Program Quesitons**

In general, when considering a graduate program, look at the following: 
*   How long is the program? Is it full-time only or is there a part-time option? Can you complete it online if that's a concern?
*   What is the cost? Are there any internal funding opportunities? 
*   What is the required curriculum? Do these required courses appeal to you or fill in your gaps in knowledge?
*   How many electives can you take? Must they be in a paritcular field (such as statistics and computer science), or is there an option to take electives in other fields as well? Do any of the electives apply to you? 
*   Are there any particular professors you would like to work with or study under? 
*   As these programs are very new, it's hard to find rankings for them. Rankings for statistics and computer science departments can be a decent proxy as many of your classes will be within these departments. 
*   What is the average or median salary for graduates' first job after graduation? What are the job placement statistics within the first year after graduation? 
*   Does the program offer substantial career help and networking opportunities? This includes interview practices, resume critiques, access to recruiters, job fairs, and more. 
*   What is the acceptance rate? What are the average GPA and GRE scores for accepted students? Is the GRE required?
*   Don't forget the imporance about location if you will be studying on campus!

#### Application Materials

Most graduate programs will ask for the following materials:
*   Undergraduate transcripts
*   Personal Statement / Statement of Purpose
*   3 letters of recommendation
*   Resume / CV
*   GRE Scores
*   TOEFL (international students only)


**Personal Statement**  
The personal statement / statement of purpose is your chance to show your genuine interest in the field and demonstrate how training at that specific instiution can help you achieve your career goals. 

Start your writing early, and do not be afraid to go through multiple drafts!

[Writing the Statement of Purpose](https://grad.berkeley.edu/admissions/apply/statement-purpose/)


**Letters of Recommendations**  
Schools will have different requirements for these letters. Some might require that at least one letter come from an academic source such as a professor. Others might be more open. In general, you want your letters to be written by the people who can best speak to your work and academic performance. Some, but not all, programs will give more weight to letters written by professors over those written by adjunct lecturers or teaching assistents. 


**Resume / CV**  
Some schools will explicitly require one or the other, so it's good to [know the difference](https://icc.ucdavis.edu/materials/resume/resumecv). 


**GRE Scores**  
Because of COVID-19, many schools are not requiring or accepting GRE scores for applications for the 2020-2021 school year. This might be a great benefit if you feel that standardized test scores do not accurately reflect your academic potential or to reduce costs. (The test is not cheap; it costs \$205 to take.)

Other schools like University of Washington have stopped requiring the GRE altogether. 

Generally, you will want your GRE quantitative score to be fairly high, but don't discount the impression a great verbal score can make. A "good" GRE score will very much depend on the program, so be sure to check the program website for the average scores of accepted students. But remember that this is just an average, and that a below average score can be offset by the strengths of your other application materials.

### Types of Interviews
Interview formats can vary by company and industry, but here are 4 common interview types:

* Behavioral Interview - the most common type of interview, consisting of “tell me about yourself” questions and examples of professional skills and prior work experiences

* Case Interview - started in the consulting world, but increasingly common in tech, finance, startups, etc., the case interview is an interactive business problem solving session

* Take Home Interview - becoming increasingly popular, and can range from programming challenges to math exercises or pre-recorded video interviews

* Coding Interview - most common in the tech/engineering world, can consist of questions that involve looking at diagrams, analyzing data, and/or assessing hypothetical problems common in the workplace

As you move through the hiring process for a job, you may have multiple types of interviews. For example, many Data Science roles require both behavioral and coding interviews at different stages in the application process.

Job interviews are an opportunity for you to ask questions about a company and evaluate whether its the right fit for you. You should feel free to ask the interviewer questions about the company, the position, or their career in the firm.
